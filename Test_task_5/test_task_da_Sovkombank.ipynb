## Задание
Нужно выполнить преобразование исходной таблицы таки образом чтобы в каждой строке были данные по каждой оказанной услуге для каждого клиента (отдельная услуга на отдельной строке).

```python
# Импорт библиотек
import pandas as pd
```

```python
# Считывание таблицы из файла
df = pd.read_excel('задание_1_реестр.xlsx', sheet_name='Исходные данные')
display(df.info())
display(df)
```

**Замечание:** Нужные данные по клиентам и услугам находятся на нескольких строках. Можно заметить определенную закономерность росположения данных.   
Поняв её, выполним преобразования таблицы таким образом чтобы вся необходимая информация по каждой услуге находилась на одной строке.

```python
# Заполнение пропущенных данных в столбце (предидущими значениями)
df['№'] = df['№'].fillna(method='ffill')
# Выделение даты оказания услуги в отдельный столбец по условию
df['Дата оказания услуги'] = df[df['Диагноз'].notna() & df['Код \nуслуги'].isna()]['Диагноз']
# Заполнение пропущенных данных в столбце (предидущими значениями)
df['Дата оказания услуги'] = df['Дата оказания услуги'].fillna(method='ffill')
# Удаление неинформативных столбцов из таблицы
df.drop(['Unnamed: 4', 'Unnamed: 5'], axis=1, inplace=True, errors='ignore')
# Удаление строк с пропущенными данными
df.dropna(inplace=True)
df
```

**Замечание:** Все необходимые данные по каждой услуге находятся на одной строке.  
Далее нужно привести таблицу к требуемому виду.

```python
# Выделение ФИО клиента в отдельный столбец
df['ФИО'] = df['№'].str.split(' ').apply(lambda x: ' '.join(x[:3]))
# Выделение Номер полиса клиента в отдельный столбец
df['Номер полиса'] = df['№'].str.split(' ').apply(lambda x: x[-1])
# Переупорядочивание колонок таблицы
df = df[['ФИО',
         'Номер полиса',
         'Дата оказания услуги',
         'Код \nуслуги',
         'Диагноз',
         'Название услуги',
         'Кол-\nво',
         'Цена по прейскуранту',
         'Сумма со скидкой'
        ]].reset_index(drop=True)
# Переименовывание колонок таблицы
df.rename(columns={"Код \nуслуги": "Код услуги",
                   "Название услуги": "Наименование услуги",
                   "Кол-\nво": "Кол-во",
                  }, inplace=True)
df
```

**Замечание:** Таблица приведена к требуемому виду.  
При необходимости можно изменить форматы данных (например признак 'Дата оказания услуги' - в формат даты).  
Далее сохраним её в эксель файл в ту же директорию.

```python
# Сохранение таблицы в эксель файл с помощью контексного менеджера
with pd.ExcelWriter('output_1.xlsx', mode='w') as writer:  
    df.to_excel(writer, sheet_name='Задание_1_Вывод', index=False)     
    for column in df:
        column_length = max(df[column].astype(str).map(len).max()*1.05, len(column))
        col_idx = df.columns.get_loc(column)
        writer.sheets['Задание_1_Вывод'].set_column(col_idx, col_idx, column_length)
```

## Задание
К исходной таблице нужно добавить 5 колонок (индекс, населенный пункт, улицу, дом и корпус). Информацию нужно выделить из адреса организации.

```python
# Импорт библиотек
import pandas as pd
import re
```

```python
# Считывание таблицы из файла
data = pd.read_excel('задание_2_тендер.xlsx', sheet_name='Исходные данные')
display(data.info())
display(data)
```

**Замечание:** Для того чтобы из адреса выделить нужную информацию напишем функцию с помощью регулярных выражений.

```python
# Функция разбивки адреса (на выходе получаем словарь с необходимой информацией)
def explore(address):
    # Создание словаря
    d = dict()
    # Выделение индекса из адреса
    d['postal_code'] = re.match(r'(\d+)', address)
    # Выделение города из адреса
    d['city'] = re.search(r'г\.(.*?),', address)
    # Выделение улицы из адреса (сначала ищем проспект по шаблону если не находим то ищем улицу)
    if re.search(r'пр-кт (\d+-[А-Яа-я\s]+),', address):
        d['street'] = re.search(r'пр-кт (\d+-[А-Яа-я\s]+),', address)
    else:
        d['street'] = re.search(r'ул\.(.*?),', address)
    # Выделение номера дома из адреса (вместе с буквой при наличии)
    d['house_number'] = re.search(r'д\.(\d+[а-я]?)', address)
    # Выделение номера корпуса дома из адреса
    d['building'] = re.search(r'корп\.(\d+)$', address)  
    
    return d
```

```python
# Добавление в таблицу необходимых столбцов с данными
data['Индекс'] = data['Адрес'].apply(lambda x: explore(x)['postal_code'].group(1) if explore(x)['postal_code'] else None)
data['Населенный пункт'] = data['Адрес'].apply(lambda x: explore(x)['city'].group(1) if explore(x)['city'] else None)
data['Улица'] = data['Адрес'].apply(lambda x: explore(x)['street'].group(1) if explore(x)['street'] else None)
data['Дом'] = data['Адрес'].apply(lambda x: explore(x)['house_number'].group(1) if explore(x)['house_number'] else None)
data['Корпус'] = data['Адрес'].apply(lambda x: explore(x)['building'].group(1) if explore(x)['building'] else None)
data
```

**Замечание:** Таблица приведена к требуемому виду.   
Далее сохраним её в эксель файл в ту же директорию. При этом будем сразу устанавливать формат колонок в эксель.

```python
# Сохранение таблицы в эксель файл с помощью контексного менеджера
with pd.ExcelWriter('output_2.xlsx', mode='w') as writer:  
    data.to_excel(writer, sheet_name='Задание_2_Вывод', index=False)     
    for column in data:
        column_length = max(data[column].astype(str).map(len).max()*1.1, len(column))
        col_idx = data.columns.get_loc(column)
        writer.sheets['Задание_2_Вывод'].set_column(col_idx, col_idx, column_length)
    writer.sheets['Задание_2_Вывод'].set_column(0, 0, 60, cell_format=writer.book.add_format({'text_wrap': True}))
        
```
