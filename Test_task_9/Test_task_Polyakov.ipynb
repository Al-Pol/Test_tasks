## Транспорт ПИК.

```python
# Импорт стандартных библиотек
import warnings

# Импорт сторонних библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Отключение ограничения вывода количества строк на экран
pd.set_option('display.max_rows', None)

# Отключение предупреждений
pd.options.mode.chained_assignment = None
warnings.filterwarnings('ignore')
```

```python
# Считываем данные из файла в датафрейм и сохраняем в переменную df
df = pd.read_excel('тест_транспорт_пик.xlsx', sheet_name='транспорт_пик')
df.head()
```

```python
# Сокращение записи колонки для удобства
df.rename(columns={
'Наименование филиала / отделения \
/ отдела или подразделения ЦПП - запрашивающего транспортировку / усиление или высвобождение транспорта':
'Наименование филиала / отделения / отдела или подразделения ЦПП'},
          inplace=True)
```

```python
df.info()
```

**Замечание:**  Рассмотрим строки с пропущенными значениями в столбце `Время начала по факту`.

```python
print('Количество пропусков:', len(df[df['Время начала по факту'].isna()]))
```

**Замечание:**  Нашлось 18 строк с пропущенными данными.   
Далее проверим данные на адекватность: соответствие даты и времени инициализации завки, начала работой над заявкой и завершение выполнения работ.

```python
# Получение полной даты завершения выполнения заявки
df['Дата и время окончания по факту'] = pd.to_datetime(df['Дата завершения заявки'].astype('str') 
                                               + ' ' 
                                               + df['Время окончания по факту'].astype('str'))
```

```python
# Рассчет даты и времени начала выполнения заявки
df['Дата и время начала по факту'] = (df['Дата и время окончания по факту'] 
                                      - pd.to_timedelta(df['Продолжительность'].astype('str')))
```

```python
# Сравнение рассчетного и заполненного времени начала выполнения заявки
print('Количество конфликтующих строк в данных:',
      len(df[df['Дата и время начала по факту'].dt.time != df['Время начала по факту']]))

df[df['Дата и время начала по факту'].dt.time != df['Время начала по факту']]
```

**Замечание:**  Нашлось 22 строки с несоответствующими логике значениями. При этом в их число вошли все 18 строк с пропущенными данными.
Далее сначала удалим эти строки и продолжим проверку данных на соответствие логике.

```python
df = df[df['Дата и время начала по факту'].dt.time == df['Время начала по факту']]
```

```python
df.info()
```

```python
print('Количество конфликтующих строк в данных:', len(df[df['Дата и время начала по факту'] < df['Дата подачи заявки']]))
```

**Замечание:** Мы нашли еще 198 строки с несоответствующими логике данными. Конечно, лучше поппытаться выяснить причину несоответствий.
Однако, далее также избавимся от этих строк.

```python
df = df[df['Дата и время начала по факту'] >= df['Дата подачи заявки']]
```

```python
# Вывод основных статистик о данных
df.describe()
```

```python
# Проверка данных на наличие дубликатов
df.duplicated().sum()
```

### Определим количество заявок по типам транспорта (по месяцам).

```python
# Вывод количества заявок по типам трпнспорта
pd.DataFrame(df['Необходимый тип транспортного средства'].value_counts())
```

```python
# Вывод количества заявок по месяцам с сортировкой по порядку в календаре
dff = df['Дата подачи заявки'].sort_values().dt.month_name().value_counts()
new_order = ['January',
             'February',
             'March',
             'April',
             'May',
             'June',
             'July',
             'August',
             'September',
             'October',
             'November', 
             'December']

dff = dff.reindex(new_order, axis=0)
pd.DataFrame(dff[dff.notna()])
```

### Определим среднюю продолжительность выполнения заявки в течении недели для каждого вида транспорта (помесячно). Представим сравнение между месяцами.

```python
# Создание дополнительных колонок дат и времени
df['Неделя выполнения заявки'] = df['Дата и время начала по факту'].dt.week
df['Месяц выполнения заявки'] = df['Дата и время начала по факту'].dt. month_name()
df['Продолжительность в минутах'] = pd.to_timedelta(df['Продолжительность'].astype('str')).dt.seconds / 60
```

```python
# Вывод средних недельных значений по типам транспорта
pd.DataFrame(df.groupby(['Неделя выполнения заявки',
                         'Необходимый тип транспортного средства'])['Продолжительность в минутах'].mean())
```

```python
# Вывод средних месячных значений по типам транспорта
pd.DataFrame(df.groupby(['Месяц выполнения заявки',
                         'Необходимый тип транспортного средства'])['Продолжительность в минутах'].mean())
```

```python
df[df['Месяц выполнения заявки'] == 'December']
```

**Замечание:** Нашлась одна отмененная заявка которая была оформлена в ноябре, а "законыена" в декабре 2022 года.

### Определим день, в который наблюдалось максимальное количество заявок (по каждому типу транспорта отдельно).

```python
# Перевод столбца в формат даты 
df['Только дата подачи заявки'] = df['Дата подачи заявки'].dt.date
```

```python
# Создание вспомогательной таблицы
data = (pd.DataFrame(df.groupby(['Необходимый тип транспортного средства',
                                'Только дата подачи заявки'])['Продолжительность'].count())
        .reset_index()
        .sort_values(by=['Необходимый тип транспортного средства',
                         'Продолжительность'], ascending=False))
data
```

```python
# Таблица с днями с максимальным количеством заявок
(data.groupby(['Необходимый тип транспортного средства'])['Продолжительность',
                                                          'Только дата подачи заявки'].first()
 .rename(columns={'Продолжительность':'Количество заявок'}))
```

### Определим суммарное количество заявок (по часам) для машин с носилками (для каждого месяца).

```python
# Получение данных только для машин с носилками
df_car = df[df['Необходимый тип транспортного средства'] == 'Машина с носилками']
# Создание дополнительных столбцов месяца и часа подачи заявок
df_car['Час подачи заявки'] = df_car['Дата подачи заявки'].dt.hour
df_car['Месяц подачи заявки'] = df_car['Дата подачи заявки'].dt.month_name()
```

```python
# Вывод количества заявок по месяцам и часам для машин с носилками
pd.DataFrame(df_car.groupby(['Месяц подачи заявки', 'Час подачи заявки'])['Продолжительность'].count())
```

## Гугл

**Замечание:** Для чтения файлов `google sheets` заменим часть ссылки: `edit?gid=0#gid=0` на `export?format=csv`.

```python
# Чтение таблиц по ссылкам и запись данных в переменные
df1 = pd.read_csv('https://docs.google.com/spreadsheets/d/1EQOYN_1kan4L45lFuXITTvxM783egVHuuFALt_M2Pec/export?format=csv')
df2 = pd.read_csv('https://docs.google.com/spreadsheets/d/1N1aKHSLSbvRhzdexJvgn6jJ35lpYVO5KtXPUSrH5umE/export?format=csv')
```

```python
df1.info()
```

```python
df2.info()
```

**Замечание:** Для объединения таблиц переименуем колонки единообразно и удалим лишние.

```python
# Переименование названий колонок
df1.columns = ['№', 'ФИО', 'Отделение']
df2.columns = ['№', 'ФИО', 'Отделение']
# Удаление колонок с номерами записей
df1 = df1.drop('№', axis=1)
df2 = df2.drop('№', axis=1)
```

```python
# Объединение таблиц в общую
result = pd.concat([df1, df2])
result.sample(10)
```

```python
# Приведение значений таблицы к одному стилю
result['Отделение'] = result['Отделение'].str.upper()
result['ФИО'] = result['ФИО'].str.capitalize()
```

```python
# Вывод уникальных значений
result['ФИО'].unique()
```

```python
# Вывод уникальных значений
result['Отделение'].unique()
```

```python
# Вывод строк с пропущенными значениями
result[result['ФИО'].isna() | result['Отделение'].isna()]
```

```python
# Удаление пропусков
print('Количество строк до удаления пропусков:', len(result))
result.dropna(inplace=True)
print('Количество строк после удаления пропусков:', len(result))
```

### Создадим сводную таблицу по всем пациентам: количество пофилиально.

```python
pd.DataFrame(result.groupby('Отделение')['ФИО'].count())
```

```python
# Сохранение таблицы в формат csv
pd.DataFrame(result.groupby('Отделение')['ФИО'].count()).to_csv('Отделения.csv')
```

### Отберем только пациентов "ПМХ".

```python
pd.DataFrame(result.loc[result['Отделение'] == 'ПМХ','ФИО'])
```

```python
# Сохранение таблицы в формат csv
pd.DataFrame(result.loc[result['Отделение'] == 'ПМХ','ФИО']).to_csv('ПМХ.csv')
```
