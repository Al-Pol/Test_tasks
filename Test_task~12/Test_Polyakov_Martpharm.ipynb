### Заполнение пропусков в столбцах `Должность` и `Город` по `ID должности`.

```python
# Создание словаря с данными по которым можно заполнять пропуски
position_cities_dict = (employees[employees['Должность'].notna()][['ID должности',
                                                                   'Должность',
                                                                   'Город']].set_index('ID должности').to_dict())
```

```python
# Заполнение пропущенных значений по "ID должности"
employees['Должность'] = employees['Должность'].fillna(employees['ID должности'].map(position_cities_dict['Должность']))
employees['Город'] = employees['Город'].fillna(employees['ID должности'].map(position_cities_dict['Город']))
```

```python
#employees['Д'] = employees.groupby('ID должности')['Должность'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
```

```python
# Вывод не заполенных данных
employees[employees['Город'].isna()]
```

```python
# Вывод не заполенных данных
employees[employees['Должность'].isna()]
```

**Замечание:**  Заметим, что данные в этих двух колонках пропущены в одних и тех же записях. 

### Заполнение строк с новыми штатными единицами, на которые еще не распределены должности и сотрудники. В графах Табельного номера и Должности проставим признак «Резерв».

**Замечание:**  Так как количество пропущенных данных одновременно в двух столбцах и по отдельности одинаковое, то можно сделать вывод, что в записях с пропущенным "табельным номером", также не заполено и "ФИО" и наоборот.

```python
# Строки с нераспределенными должностями и сотрудниками
employees[employees['ФИО'].isna() & employees['Должность'].isna()]
```

```python
# Перевод данных в тип str для того, чтобы можно было проставить признак "Резерв"
employees['Табельный номер'] = employees['Табельный номер'].astype('str')
# Заполнение необходимых строк записью "Резерв"
employees.loc[employees['ФИО'].isna() 
          & employees['Должность'].isna(), 'Табельный номер'] = 'Резерв'
employees.loc[employees['ФИО'].isna() 
          & employees['Должность'].isna(), 'Должность'] = 'Резерв'
```
