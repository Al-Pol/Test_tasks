## Задание

### Условие.

Написать функцию которая удаляет дубликаты из списка сохраняя его порядок. На вход ф-ции подается список, на выходе...

```python
ls = ['one', 'two', 'one', 'two', 'three', 'two', 'one', 'four', 'three', 'five']
```

### Решение.

При написании функции будем использовать аннотацию типов данных.

```python
def remove_duplicates(lst: list) -> list:
    result = []
    for item in lst:
        if item not in result:
            result.append(item)
    return result
```

```python
remove_duplicates(ls)
```

## Задание

### Условие.

- В папке Reports находятся excel файлы с выгрузками из базы данных по каждому дню, файлы имеют схожее название типа - "DAG-111_patients_2023-02-01.xlsx" имеют общую структуру, могут иметь несколько листов данных. Их нужно объединить в один ДатаСет и в нем должна содержатся колонка, в которой информация из, какого файла загружена запись

- Файлы со значениями анализов пациентов за 2022 и 2023 год. Нужно посчитать среднее значение показателей пациентов за 2023 год общее и по каждому пациенту в отдельности.

Файл имеет 2 столбца:
    id - id пациента
    values - перечисленные анализы пациента разделенные ";"

### Решение.



#### Предобоаботка.

```python
# Импорт библиотек
import pandas as pd
import glob
```

```python
# Поиск файлов в папке
folder_path = 'Reports'
files = glob.glob(folder_path + '/*.xlsx')

# Считывание данных со всех листов excel из файлов найденных в папке на предидущем шаге
# Объединение данных какждого листа из файла excel в одну таблицу и создание столбца с именем исходного файла
# Объединение получившихся таблиц в итоговую
lst_dir = []
for file in files:
    dict_file = pd.read_excel(file, sheet_name=None)
    lst_file = []
    for sheet in dict_file.keys():
        lst_file.append(dict_file[sheet])
    df_file = pd.concat(lst_file)                              
    df_file['file_name'] = file.replace('Reports\\', '')
    lst_dir.append(df_file)
df_dir = pd.concat(lst_dir)
# Сброс индексов и вывод таблицы на экран
df_dir.reset_index(drop=True) 
```

**Замечание:**
В столбце `values` представлены строки с данными разной длины.

```python
# Разбиение строки
df_dir['values'] = df_dir['values'].str.split(';')
# Вывод 2-х первых строк таблицы
df_dir.head(2)
```

**Замечание:**
В результате столбец `values` содержит список строк с данными.

```python
# Преобразование таблицы: разворачивание каждого элемента списка в столбце "values" в отдельную строку
df_dir = df_dir.explode('values', ignore_index=True)
# Выделение столбцов с датой "date" и значением анализа пациента "value" в отдельные столбцы 
df_dir[['date', 'value']] = df_dir['values'].str.split(' - ', expand=True)
# Вывод таблицы на экран
df_dir
```

**Замечание:**
В одной строке у нас пропущенны данные (`Хил` вместо числа). Заметим, что у нас есть одна строка с похожим названием `67,00 (Хилезная сыворотка)`. Предположим, что `Хил` = `Хилезная сыворотка`, поэтому заменим значение `Хил` числом `67`.  

Далее обработаем строки и приведем значения в столбцах к соответствующим типам данных.

```python
# Замена "пропущенного" значения
df_dir.loc[df_dir['value'] == 'Хил', 'value'] = '67'

# Удаление буквенных символов и круглых скобок и замена запятых на точки
df_dir['value'] = (df_dir['value'].str.replace('[A-Za-zА-Яа-я]+', '', regex=True)
                                  .str.replace('\(.*\)', '', regex=True)
                                  .str.replace(',', '.', regex=True))

# Приведение столбца к числовому типу "float"
df_dir['value'] = pd.to_numeric(df_dir['value'])

# Приведение столбца к числовому типу даты "datetime"
df_dir['date'] = pd.to_datetime(df_dir['date'], format='%Y-%m-%d')

# Вывод информации о датафрейме
df_dir.info()
```

####  Вывод необходимых показателей.

```python
# Выделение года из даты в отдельный столбец
df_dir['year'] = df_dir['date'].dt.year

# Фильтрация данных таблицы по году 
df_dir_2023 = df_dir[df_dir['year'] == 2023]

# Вывод искомых средних показателей
print()
print()
print(f"Среднее значение всех показателей за 2023 год: {df_dir_2023['value'].mean()}")
print('-----------------------------------------------------')
print()
print()
print('Среднее значение показателей каждого пациента за 2023 год')
df_dir_2023[['id', 'value']].groupby(['id']).mean().T
```

## Задание

### Условие.
- В папке visits файл посещения. Столбцу "Список пройденных исследований" соответствует "Список дат пройденных исследований".
-  Необходимо произвести очистку от дублей и оставить последние актуальные анализы (каждому уникальному анализу должна соответствовать последняя дата)

### Решение.

#### Предобработка.

```python
# Считывание файла и запись в переменную
visits = pd.read_excel('visits/посещения.xlsx')
# Перевод данных из строкового состояния в списковое по разделителю ";"
visits['Список пройденных исследований'] = visits['Список пройденных исследований'].str.split(';')
visits['Список дат пройденных исследований'] = visits['Список дат пройденных исследований'].str.split(';')
visits
```

**Замечание:**
По условию задачи: `Столбцу "Список пройденных исследований" соответствует "Список дат пройденных исследований"`.    
Это соответствие важно, для того, чтобы мы могли развернуть данные списки в строки для удобства работы. Проверим длины получившихся списков.

```python
# Получение колонок с длинами списков
visits['len_research'] = visits['Список пройденных исследований'].str.len()
visits['len_date'] = visits['Список дат пройденных исследований'].str.len()
visits
```

**Замечание:**
Мы видим несоответсвие для 2х строк. Количество анализов (10) больше количества дат соответствующих данным анализам (9).
В такой ситуации у нас не получится развернуть списки в строки (без искажения данных). Нужно проверить эти строки (4 и 6).

```python
visits['Список пройденных исследований'][4]
```

```python
visits['Список пройденных исследований'][6]
```

**Замечание:**
Заметим, что значению `микроскопическое исследование мазка крови при выявлении патологии` нет соответствующего кодового обозначения. Возможно эта запись относится к `ESLI.LI.C.2.2 - Общий клинический анализ крови (общий анализ + СОЭ)`. 

Однако после небольшлго изучения предметной области выяснилось, что эти исследования отличаются. Скорее всего, эта запись образовалась по ошибке (или её забыли удалить). Конечно, лучше эту информацию уточнить у заказчика. В данном случае, просто удалим эти записи (в других строках эти записи не встречаются).

```python
# Удаление строк
try:
    visits['Список пройденных исследований'][4].remove(' микроскопическое исследование мазка крови при выявлении патологии')
    visits['Список пройденных исследований'][6].remove(' микроскопическое исследование мазка крови при выявлении патологии')
except:
    pass    
```

```python
# Преобразование таблицы: разворачивание каждого элемента списка в столбцах "списков" в отдельную строку
visits = visits.explode(['Список пройденных исследований', 'Список дат пройденных исследований'], ignore_index=True)

# Выделение столбцов с кодом "code" и названием исследования "research" в отдельные столбцы 
visits[['code', 'research_name']] = visits['Список пройденных исследований'].str.split(' - ', expand=True)

# Удаление пробелов в начале и в конце строк
visits['code'] = visits['code'].str.strip()
visits['research_name'] = visits['research_name'].str.strip()
```

```python
# Удаление и переименование столбцов
visits = visits.drop(['len_research', 'len_date', 'Список пройденных исследований'], axis=1, errors='ignore')
visits = visits.rename(columns={'Список дат пройденных исследований': 'date'})

# Приведение столбца к числовому типу даты "datetime"
visits['date'] = pd.to_datetime(visits['date'], format='%Y-%m-%d')

# Вывод таблицы
visits
```

**Замечание:**
В итоге получилась таблица длиной 46 строк.

#### Обработка дубликатов.
По условию задачи таблицу нужно очистить от дубликатов. Кажому уникальному виду исследования соответствует единственный код, поэтому далее будем анализировать анализы по колонке `code`.

```python
# Поиск дубликатов в колонках "списков"
print(f"Обнаружено дубликатов: {visits[['ИД пациента', 'code', 'date']].duplicated().sum()}")
```

```python
# Удаление дубликатов
visits = visits.drop_duplicates(subset=['ИД пациента', 'code', 'date'], keep='first')
print(f'Размер получившейся таблицы: {visits.shape}')
```

**Замечание:**
Удалили одну строку - дубликат.

#### Актуализация анализов.
По условию задачи в итоговой таблице нужно оставить последние актуальные анализы (каждому уникальному анализу должна соответствовать последняя дата).

Решая задачу примем, что данную проверку нужно проводить для каждого пациента (`id`) отдельно.

```python
# Фильтрация данных
result = visits[visits['date'] == visits.groupby(['ИД пациента', 'code'])['date'].transform('max')]
result
```

Решая эту задачу мы сгруппировали данные по `id` пациента и коду анализа, после этого оставили только последние даты. Затем в таблице оставили строки соответствующие этим датам.

В итоге мы отфильтровали одну строку - неактуальный анализ.

## Задание

### Условие.

- В папке visits файл посещения.
- В папке visits файл словарь.

- Необходимо произвести категоризацию диагнозов из столбца "Код мкб-10" файла посещения:
  - Создать словарь из файла словарь, колонка "Группа ДН" - ключ, колонка "kod_mkb_10" - значения.
  - Вывести результат в новом столбце("dn") файла посещения - без дублей, сортированный

### Решение.

```python
# Считывание файла и запись в переменную
df_dict = pd.read_excel('visits/словарь.xlsx')
# Считывание файла и запись в переменную
df_visits = pd.read_excel('visits/посещения.xlsx')
```

#### Создание словаря.

Так ключи в словаре должны быть уникальными, то создадим словарь в котором:   
ключи - колонка `kod_mkb_10`, значения - `Группа ДН`.

```python
# Создание словаря
dct = {}
for i, row in df_dict.iterrows():
    dct[row['kod_mkb_10']] = row['Группа ДН']

dct
```

#### Категоризация диагнозов.

```python
# Создание столбца с категориями диагнозов взятыми из словаря dct
df_visits['dn'] = df_visits['Код мкб-10'].apply(lambda x: ';'.join(sorted(set([dct[i] for i in x.split(';')]))))
df_visits
```

**Замечание:**
Мы создали столбец `dn`, в котором указаны категории диагнозов. При этом если диагнозов несколько (записаны через ";"), то в столбце `dn` каждому диагнозу будет соответствовать своя категория (записаны также через ";").
