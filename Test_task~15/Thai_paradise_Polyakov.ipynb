## Общая информация.

```python
# Импорт стандартных библиотек
import warnings

# Импорт сторонних библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Отключение предупреждений
pd.options.mode.chained_assignment = None
warnings.filterwarnings('ignore')
```

### Считываем данные из CSV-файла в датафрейм и сохраняем в переменную df.

```python
df = pd.read_csv('visits_test2.csv')
```

```python
df.sample(10)
```

## Преобразование и очистка данных.

### Преобразуем типы данных.

```python
# Преобразование столбцов в формат даты
df['Дата и время'] = pd.to_datetime(df['Дата и время'], format='%Y-%m-%d %H:%M:%S')
df['Дата первого визита'] = pd.to_datetime(df['Дата первого визита'], format='%Y-%m-%d %H:%M:%S')
# Преобразование столбца в формат числа с дробного числа
df['Количество часов'] = df['Количество часов'].str.replace(',','.').astype(float)
df.info()
```

### Проверим датафрейм на наличие явных дубликатов.

```python
# Поиск дубликатов
df.duplicated().sum()
```

**Замечание:**  Нашлось 1009 дублирующих записей.

```python
# Удаление дубликатов
df = df.drop_duplicates()
```

**Замечание:**  Удалили дубликаты.

### Обработаем пропуски в данных.

```python
# Вывод количества пропущенных данных
df.isna().sum()
```

**Замечание:**  В колонке `Вид платежного документа` есть пропущенные значения (`NaN`). Это может быть связано с тем, что для некоторых способов оплаты (например, `Нал/Безнал` или `Прочее`) не требуется указывать вид платежного документа.

```python
# Создание копии датасета
df_filled = df.copy()
```

```python
# Заполнение пропусков заглушкой
df_filled = df.fillna('Не указано')
```

**Замечание:**  Заполнили пропуски в исходном датасете в данных заглушкой `Не указано`.

### Поиск аномалий.

```python
# Поиск строк с аномальными данными даты и времени
len(df_filled[df_filled['Дата и время'] <  df_filled['Дата первого визита']])
```

 **Замечание:**  `Дата первого визита` всегда идет до `Дата и время`. Нестыковок не обнаружено.

```python
# Проверка данных на аномаьлность
df_filled[df_filled['Количество часов'] <= 0].sample(5)
```

 **Замечание:**  В колонке `Количество часов` есть значения `0`, что может быть аномалией, так как продолжительность услуги не может быть нулевой. Также в ходе анализа сайта компании удалось выяснить, что минимальная длительность услуги составляет `30 минут`, поэтому отфильтруем сомнительные данные.

```python
# Удаление строк со значениями меньше 30 минут
df_filled = df_filled[df_filled['Количество часов'] >= 0.5]
```

 **Замечание:**  Удалили строки с нулевой продолжительностью услуги.

```python
# Поиск дубликатов по полю "id"
print('Количество дубликатов по полю id:', len(df_filled[df_filled['id'].duplicated()]))
# Вывод строк с дубликатами
df_filled[df_filled['id'].duplicated(keep=False)].head(10)
```

 **Замечание:**  В данных есть дубликаты по полю `id`. Это может быть ошибкой в данных, так как `id` визита должен быть уникальным. C другой стороны данные в других колонка разнятся. Поэтому прежде, чем удалять эти строки нужно выяснить причину дублирования. Например, услуга может быть парной: два клиента одновременно.  
**Оставим данные строки в таблице.**

## Визуализация.
Построим графики этих показателей по месяцам.

```python
# Создание колонки месяца по годам
df_filled['Месяц'] = df['Дата и время'].dt.to_period('M')
```

### Средняя продолжительность услуг по месяцам.

```python
# Группировка данных
monthly_avg_duration = df_filled.groupby('Месяц')['Количество часов'].mean()

# Построение графика
plt.figure(figsize=(10, 6))
monthly_avg_duration.plot(kind='bar')
plt.title('Средняя продолжительность услуг по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Средняя продолжительность (часы)')
plt.show()
```

### Количество уникальных клиентов по месяцам.

```python
# Группировка данных
monthly_unique_clients = df_filled.groupby('Месяц')['Клиент'].nunique()

# Построение графика
plt.figure(figsize=(10, 6))
monthly_unique_clients.plot(kind='bar')
plt.title('Количество уникальных клиентов по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Количество уникальных клиентов')
plt.show()
```

### Количество новых клиентов по месяцам.

```python
# Поиск клиенты с одной записью
visits = df_filled.groupby(['Клиент'])['id'].count().reset_index()
clients = visits[visits['id'] == 1]['Клиент'].to_list()

# Получение таблицы с новыми клиентами
new_clients = df_filled[df_filled['Клиент'].isin(clients) 
                       & (df_filled['Дата и время'] == df_filled['Дата первого визита'])]

# Группировка данных
new_clients_monthly = new_clients.groupby('Месяц')['Клиент'].nunique()

# Построение графика
plt.figure(figsize=(10, 6))
new_clients_monthly.plot(kind='bar')
plt.title('Количество новых клиентов по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Количество новых клиентов')
plt.show()
```
