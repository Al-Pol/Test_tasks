## Выявление закономерностей.

###  Найдём корреляцию между количеством звонков и количеством встреч.

```python
# Корреляция между количеством звонков и количеством встреч
df[['Звонки', 'Встречи']].corr()
```

**Замечание:** Корреляция показывает сильно выраженную линейную связь между количеством звонков и встреч.   
Для наглядности визуализируем эту связь.

```python
# Построение точечного графика с регрессионной линией
sns.regplot(data=df, x='Звонки', y='Встречи')
plt.title('Зависимость количества Встреч от количества Звонков')
plt.xlabel('Количество звонков (исходящих)')
plt.ylabel('Количество встреч с клиентами');
```

###  Определим, есть ли связь между количеством SMS и количеством подписанных договоров.

```python
# Корреляция между количеством SMS и количеством подписанных договоров
df[['SMS', 'Договора']].corr()
```

**Замечание:** Корреляция показывает сильно выраженную линейную связь между количеством SMS и подписанных договоров.

```python
# Построение точечного графика с регрессионной линией
sns.regplot(data=df, x='SMS', y='Договора')
plt.title('Зависимость количества SMS и подписанных договоров')
plt.xlabel('Количество SMS (отправленных клиентам)')
plt.ylabel('Количество подписанных договоров с клиентами');
```

###  Проверим, влияет ли среднее время подготовки листинга на количество завершенных сделок.

```python
# Корреляция между средним временем подготовки листинга и количеством завершенных сделок
df[['Ср. листинг (часы)','Сделки']].corr()
```

**Замечание:** Корреляция не показывает средним временем подготовки листинга и количеством завершенных сделок.   
Однако связь между признаками может быть нелинейной.
Для анализа визуализируем эту связь.

```python
# Построение точечного графика
plt.scatter(df['Ср. листинг (часы)'], df['Сделки'])
plt.title('Зависимость количества сделок от времени подготовки листинга')
plt.xlabel('Среднее время подготовки листинга (в часах)')
plt.ylabel('Количество завершенных сделок');
```

**Замечание:** Здесь мы не можем увидеть мвязи между признаками.  
Однако связь между ними все же может быть. Возможно нам не хватает данных или признаки связаны между собой через какие-либо другие. Например время подготовки листинга влияет на его количество, а количество листингов влияет на количество сделок. И таким образом эти признаки связаны.

```python
# Построение точечного графика
plt.scatter(df['Ср. листинг (часы)'], df['Листинги'])
plt.title('Зависимость количества листингов \n от времени подготовки листинга')
plt.xlabel('Среднее время подготовки листинга (в часах)')
plt.ylabel('Количество выставленных листингов \n (объектов недвижимости)');
```

```python
# Построение графиков зависимости суммы и среднего времени подготовки листинга (в часах) от количества сделок
pd.DataFrame(df.groupby('Сделки')['Ср. листинг (часы)'].agg(['mean', 'sum'])).plot(kind='bar')
plt.xticks(rotation=0)
plt.legend(['Среднее', 'Сумма'])
plt.title('Зависимость суммы и среднего времени \n подготовки листинга (в часах) от количества сделок')
plt.xlabel('Количество завершенных сделок')
plt.ylabel('Среднее время подготовки листинга (в часах)');
```

**Замечание:** В данном случае связь выявить не удалось. Однако одназначно утверждать это преждевременно. Признаки могут быть связаны другим способом.

```python
# Построение точечного графика
plt.scatter(df['Ср. листинг (часы)'], df['Сделки'])
plt.title('Зависимость количества сделок от времени подготовки листинга')
plt.xlabel('Среднее время подготовки листинга (в часах)')
plt.ylabel('Количество завершенных сделок');
```
