### Проверим как взаимосвязаны признаки с помощью корреляции.

С помощью корреляции `Пирсона` можно оценить числовые признаки, но для категориальных более подходит метод `phik`.

```python
# Функция вывода таблицы корреляции в виде тепловой карты
def corr_plot(data, corr_type):
    '''
    data - таблица с данными
    corr_type - тип корреляции ('pirson' или 'phik')
    '''
    if corr_type == 'pirson':
        table = data.corr()
    if corr_type == 'phik':
        table = data.phik_matrix(verbose=False)
    #mask = np.triu(table, k=1);
    plt.figure(figsize=(20, 15));
    sns.heatmap(table,
                square=True,
                annot=True,
                linewidth=0.5,
                cmap='coolwarm',
                #mask=mask
               ); 
```

```python
# Вывод корреляции для исходных данных
print(corr_plot(df, 'pirson'))
print(corr_plot(df, 'phik'))
```

```python
# Вывод корреляции для очищенных данных
print(corr_plot(df_clear, 'pirson'))
print(corr_plot(df_clear, 'phik'))
```

 **Замечание:**  На графиках видно, что даты регистрации и последнего прослушивания треков с последним входом коррелируют. При этом на "очищенных" данных корреляция более выражена.

## Посмотрим на  количество дней `playtrack` для разных диапазонов даты первого входа.

```python
seasons = [1, 1, 2, 2, 2, 3, 3, 3, 4, 4, 4, 1]
month_to_season = dict(zip(range(1,13), seasons))
df['season'] = df['first_login_dt'].dt.month.map(month_to_season) 
df['year'] = df['first_login_dt'].dt.year
df.groupby(['year','season']).count()['first_login_dt']
```

```python
df.groupby(['year','season']).count()['first_login_dt'].plot.bar()
plt.xlabel('Год и сезон')
plt.ylabel('Количество регистраций');
```

**Замечание:** На графике видно количество регистраций пользователей разбитое по годам и сезонам. Больше всего регистраций было в 4 квартале 2022 года.
